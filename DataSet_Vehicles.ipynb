# Análisis del Vehicle Dataset - Regresión Lineal

## Descripción
Análisis completo del dataset de vehículos de CardekHo, que incluye:
- car_data.csv
- Car_Details.csv
- Car_Details_V3.csv
- Car_Details_V4.csv

## Objetivo
Predecir el precio del automóvil en función de sus atributos usando regresión lineal.

## Estructura del Análisis
1. Carga y unificación de datos
2. Análisis exploratorio
3. Preprocesamiento
4. Modelado usando regresión lineal
5. Evaluación de resultados

```python
# Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# Configuración de visualización
plt.style.use('default')
%matplotlib inline
sns.set_theme(style="whitegrid")

# 1. Carga de los diferentes archivos del Vehicle Dataset
print("Cargando datasets...")
df1 = pd.read_csv('car data.csv')
df2 = pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')
df3 = pd.read_csv('Car details v3.csv')
df4 = pd.read_csv('car details v4.csv')

# Mostrar información básica de cada dataset
def mostrar_info_dataset(df, nombre):
    print(f"\nInformación de {nombre}:")
    print("-" * 50)
    print(f"Dimensiones: {df.shape}")
    print("\nColumnas:")
    print(df.columns.tolist())
    print("\nPrimeras 3 filas:")
    print(df.head(3))
    print("\nTipos de datos:")
    print(df.dtypes)

# Mostrar información de cada dataset
mostrar_info_dataset(df1, "Dataset 1 - car data.csv")
mostrar_info_dataset(df2, "Dataset 2 - CAR DETAILS FROM CAR DEKHO.csv")
mostrar_info_dataset(df3, "Dataset 3 - Car details v3.csv")
mostrar_info_dataset(df4, "Dataset 4 - car details v4.csv")

# 2. Análisis de columnas comunes y diferentes
def analizar_columnas_comunes(dataframes):
    nombres = ['df1', 'df2', 'df3', 'df4']
    columnas = [set(df.columns) for df in dataframes]
    
    print("\nAnálisis de columnas:")
    print("-" * 50)
    
    # Columnas comunes entre todos los datasets
    columnas_comunes = set.intersection(*columnas)
    print("\nColumnas comunes en todos los datasets:")
    print(sorted(list(columnas_comunes)))
    
    # Columnas únicas en cada dataset
    for i, cols in enumerate(columnas):
        cols_unicas = cols - set.union(*[c for j, c in enumerate(columnas) if j != i])
        print(f"\nColumnas únicas en {nombres[i]}:")
        print(sorted(list(cols_unicas)))

# Analizar columnas comunes y únicas
analizar_columnas_comunes([df1, df2, df3, df4])
```

```python
# Función para estandarizar nombres de columnas
def estandarizar_nombres_columnas(df, tipo_dataset):
    df_copy = df.copy()
    
    if tipo_dataset == 1:  # car data.csv
        mapping = {
            'Car_Name': 'name',
            'Year': 'year',
            'Selling_Price': 'selling_price',
            'Kms_Driven': 'km_driven',
            'Fuel_Type': 'fuel',
            'Seller_Type': 'seller_type',
            'Transmission': 'transmission',
            'Owner': 'owner'
        }
    elif tipo_dataset == 2:  # CAR DETAILS FROM CAR DEKHO.csv
        # Ya tiene los nombres estandarizados
        return df_copy
    elif tipo_dataset == 3:  # Car Details v3.csv
        # Ya tiene los nombres básicos estandarizados
        return df_copy
    elif tipo_dataset == 4:  # Car Details v4.csv
        mapping = {
            'Make': 'manufacturer',
            'Model': 'name',
            'Price': 'selling_price',
            'Year': 'year',
            'Kilometer': 'km_driven',
            'Fuel Type': 'fuel',
            'Transmission': 'transmission',
            'Owner': 'owner',
            'Seller Type': 'seller_type'
        }
    
    # Aplicar el mapeo
    columnas_a_renombrar = {k: v for k, v in mapping.items() if k in df_copy.columns}
    df_copy = df_copy.rename(columns=columnas_a_renombrar)
    return df_copy

# Estandarizar nombres en todos los datasets
df1_std = estandarizar_nombres_columnas(df1, 1)
df2_std = estandarizar_nombres_columnas(df2, 2)
df3_std = estandarizar_nombres_columnas(df3, 3)
df4_std = estandarizar_nombres_columnas(df4, 4)

# Función para estandarizar valores
def estandarizar_valores(df):
    df_copy = df.copy()
    
    # Estandarizar owner
    if 'owner' in df_copy.columns:
        owner_mapping = {
            0: 'First Owner',
            1: 'Second Owner',
            2: 'Third Owner',
            3: 'Fourth Owner',
            '0': 'First Owner',
            'First Owner': 'First Owner',
            'Second Owner': 'Second Owner',
            'Third Owner': 'Third Owner',
            'Fourth & Above Owner': 'Fourth Owner'
        }
        df_copy['owner'] = df_copy['owner'].map(owner_mapping).fillna(df_copy['owner'])
    
    return df_copy

# Estandarizar valores en todos los datasets
df1_std = estandarizar_valores(df1_std)
df2_std = estandarizar_valores(df2_std)
df3_std = estandarizar_valores(df3_std)
df4_std = estandarizar_valores(df4_std)

# Seleccionar columnas comunes principales para el análisis
columnas_principales = ['name', 'year', 'selling_price', 'km_driven', 'fuel', 
                       'seller_type', 'transmission', 'owner']

# Unificar datasets con columnas principales
df1_unified = df1_std[columnas_principales].copy()
df2_unified = df2_std[columnas_principales].copy()
df3_unified = df3_std[columnas_principales].copy()
df4_unified = df4_std[columnas_principales].copy()

# Concatenar todos los datasets
df_final = pd.concat([df1_unified, df2_unified, df3_unified, df4_unified], 
                    axis=0, ignore_index=True)

# Eliminar duplicados si existen
df_final = df_final.drop_duplicates()

# Mostrar información del dataset unificado
print("\nInformación del dataset unificado:")
print("-" * 50)
print(f"Dimensiones finales: {df_final.shape}")
print("\nPrimeras filas del dataset unificado:")
print(df_final.head())
print("\nEstadísticas descriptivas:")
print(df_final.describe())
print("\nValores nulos:")
print(df_final.isnull().sum())
```

```python
# Análisis Exploratorio de Datos (EDA)
plt.figure(figsize=(20, 15))

# 1. Distribución del precio de venta
plt.subplot(3, 2, 1)
sns.histplot(data=df_final, x='selling_price', bins=50)
plt.title('Distribución del Precio de Venta')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')

# 2. Distribución de años
plt.subplot(3, 2, 2)
sns.countplot(data=df_final, x='year')
plt.xticks(rotation=45)
plt.title('Distribución por Año')

# 3. Box plot de precio por tipo de combustible
plt.subplot(3, 2, 3)
sns.boxplot(data=df_final, x='fuel', y='selling_price')
plt.xticks(rotation=45)
plt.title('Precio por Tipo de Combustible')

# 4. Scatter plot de precio vs kilómetros
plt.subplot(3, 2, 4)
sns.scatterplot(data=df_final, x='km_driven', y='selling_price', alpha=0.5)
plt.title('Precio vs Kilómetros Recorridos')

# 5. Precio promedio por tipo de transmisión
plt.subplot(3, 2, 5)
sns.boxplot(data=df_final, x='transmission', y='selling_price')
plt.title('Precio por Tipo de Transmisión')

# 6. Precio por tipo de propietario
plt.subplot(3, 2, 6)
sns.boxplot(data=df_final, x='owner', y='selling_price')
plt.xticks(rotation=45)
plt.title('Precio por Tipo de Propietario')

plt.tight_layout()
plt.show()

# Análisis de correlaciones numéricas
plt.figure(figsize=(10, 8))
numeric_cols = ['year', 'selling_price', 'km_driven']
sns.heatmap(df_final[numeric_cols].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlación de Variables Numéricas')
plt.show()

# Estadísticas por categoría
print("\nEstadísticas de precio por tipo de combustible:")
print(df_final.groupby('fuel')['selling_price'].describe())

print("\nEstadísticas de precio por tipo de transmisión:")
print(df_final.groupby('transmission')['selling_price'].describe())

print("\nEstadísticas de precio por tipo de vendedor:")
print(df_final.groupby('seller_type')['selling_price'].describe())

# Preparación para el modelo de regresión lineal
def preparar_datos_modelo(df):
    # Crear copias para no modificar los datos originales
    df_model = df.copy()
    
    # Codificación de variables categóricas
    categorical_columns = ['fuel', 'seller_type', 'transmission', 'owner']
    for column in categorical_columns:
        dummies = pd.get_dummies(df_model[column], prefix=column, drop_first=True)
        df_model = pd.concat([df_model, dummies], axis=1)
    
    # Eliminar columnas originales categóricas y la columna 'name'
    df_model = df_model.drop(categorical_columns + ['name'], axis=1)
    
    return df_model

# Preparar datos para el modelo
df_model = preparar_datos_modelo(df_final)

print("\nColumnas preparadas para el modelo:")
print(df_model.columns.tolist())

# Guardar el dataset procesado
df_model.to_csv('vehicle_dataset_processed.csv', index=False)
print("\nDataset procesado guardado como 'vehicle_dataset_processed.csv'")
```

```python
# Importar bibliotecas adicionales necesarias
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.preprocessing import StandardScaler

# 1. Preparación de variables
X = df_model.drop('selling_price', axis=1)
y = df_model['selling_price']

# 2. Escalado de características numéricas
scaler = StandardScaler()
X_scaled = X.copy()
numeric_cols = ['year', 'km_driven']
X_scaled[numeric_cols] = scaler.fit_transform(X_scaled[numeric_cols])

# 3. Selección de características
k_best = SelectKBest(score_func=f_regression, k=10)
X_selected = k_best.fit_transform(X_scaled, y)

# Obtener las características seleccionadas
selected_features_mask = k_best.get_support()
selected_features = X_scaled.columns[selected_features_mask].tolist()

print("Características seleccionadas:")
for i, feature in enumerate(selected_features, 1):
    score = k_best.scores_[k_best.get_support()][i-1]
    print(f"{i}. {feature}: {score:.2f}")

# 4. División del dataset
X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y, test_size=0.2, random_state=42
)

print("\nDimensiones de los conjuntos de datos:")
print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_test: {y_test.shape}")

# Guardar los datos preparados para el modelo
np.save('X_train.npy', X_train)
np.save('X_test.npy', X_test)
np.save('y_train.npy', y_train)
np.save('y_test.npy', y_test)
print("\nDatos de entrenamiento y prueba guardados en archivos .npy")
```

```python
# 1. Importar bibliotecas adicionales necesarias
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

# 2. Definir y entrenar el modelo de regresión lineal con validación cruzada
param_grid = {
    'fit_intercept': [True, False],
    'positive': [True, False]
}

model = LinearRegression()
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Imprimir mejores parámetros
print("Mejores parámetros encontrados:")
print(grid_search.best_params_)

# 3. Realizar predicciones
y_pred_train = grid_search.predict(X_train)
y_pred_test = grid_search.predict(X_test)

# 4. Evaluar el modelo
def evaluar_modelo(y_true, y_pred, conjunto=""):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    
    print(f"\nMétricas de evaluación ({conjunto}):")
    print(f"Error cuadrático medio (MSE): {mse:,.2f}")
    print(f"Raíz del error cuadrático medio (RMSE): {rmse:,.2f}")
    print(f"Error absoluto medio (MAE): {mae:,.2f}")
    print(f"Coeficiente de determinación (R²): {r2:.4f}")

# Evaluar en conjunto de entrenamiento y prueba
evaluar_modelo(y_train, y_pred_train, "Entrenamiento")
evaluar_modelo(y_test, y_pred_test, "Prueba")

# 5. Visualización de resultados
plt.figure(figsize=(15, 10))

# Gráfico 1: Valores reales vs predicciones (Training)
plt.subplot(2, 2, 1)
plt.scatter(y_train, y_pred_train, alpha=0.5)
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'r--', lw=2)
plt.xlabel('Valores Reales')
plt.ylabel('Predicciones')
plt.title('Valores Reales vs Predicciones (Training)')

# Gráfico 2: Valores reales vs predicciones (Test)
plt.subplot(2, 2, 2)
plt.scatter(y_test, y_pred_test, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Valores Reales')
plt.ylabel('Predicciones')
plt.title('Valores Reales vs Predicciones (Test)')

# Gráfico 3: Distribución de residuos (Training)
plt.subplot(2, 2, 3)
residuos_train = y_train - y_pred_train
sns.histplot(residuos_train, kde=True)
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')
plt.title('Distribución de Residuos (Training)')

# Gráfico 4: Distribución de residuos (Test)
plt.subplot(2, 2, 4)
residuos_test = y_test - y_pred_test
sns.histplot(residuos_test, kde=True)
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')
plt.title('Distribución de Residuos (Test)')

plt.tight_layout()
plt.show()

# 6. Análisis de coeficientes e importancia de características
coef = grid_search.best_estimator_.coef_
feature_importance = pd.DataFrame({
    'Feature': selected_features,
    'Importance': np.abs(coef)
})
feature_importance = feature_importance.sort_values('Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importance, x='Importance', y='Feature')
plt.title('Importancia de las Características')
plt.xlabel('Magnitud del Coeficiente')
plt.tight_layout()
plt.show()

# 7. Guardar el modelo y resultados
import joblib
joblib.dump(grid_search.best_estimator_, 'modelo_regresion_lineal.pkl')
print("\nModelo guardado como 'modelo_regresion_lineal.pkl'")

# 8. Resumen final
print("\nRESUMEN DEL MODELO")
print("-" * 50)
print("1. Mejor modelo:")
print(f"   - R² (Training): {r2_score(y_train, y_pred_train):.4f}")
print(f"   - R² (Test): {r2_score(y_test, y_pred_test):.4f}")
print("\n2. Características más importantes:")
print(feature_importance.head())
print("\n3. Diagnóstico de residuos:")
print(f"   - Media de residuos (Test): {residuos_test.mean():.2f}")
print(f"   - Desviación estándar de residuos (Test): {residuos_test.std():.2f}")
```

```python
# Veamos las métricas detalladamente
print("\nANÁLISIS DE RESULTADOS:")
print("-"*50)
print("\n1. Comparación de métricas entre Training y Test:")
metricas_df = pd.DataFrame({
    'Métrica': ['R² Score', 'MSE', 'RMSE', 'MAE'],
    'Training': [r2_score(y_train, y_pred_train),
                mean_squared_error(y_train, y_pred_train),
                np.sqrt(mean_squared_error(y_train, y_pred_train)),
                mean_absolute_error(y_train, y_pred_train)],
    'Test': [r2_score(y_test, y_pred_test),
             mean_squared_error(y_test, y_pred_test),
             np.sqrt(mean_squared_error(y_test, y_pred_test)),
             mean_absolute_error(y_test, y_pred_test)]
})
print(metricas_df.round(4))
```

```python
# Análisis de características más influyentes
print("\n2. Importancia de características (Top 5):")
print(feature_importance.head().round(4))

# Visualización mejorada
plt.figure(figsize=(12, 6))
sns.barplot(data=feature_importance, x='Importance', y='Feature')
plt.title('Importancia Relativa de las Características')
plt.xlabel('Magnitud del Impacto')
plt.ylabel('Características')
plt.tight_layout()
plt.show()
```

```python
# Análisis detallado de residuos
residuos_stats = pd.DataFrame({
    'Estadística': ['Media', 'Desviación Estándar', 'Mínimo', 'Máximo'],
    'Training': [residuos_train.mean(), residuos_train.std(), 
                residuos_train.min(), residuos_train.max()],
    'Test': [residuos_test.mean(), residuos_test.std(), 
             residuos_test.min(), residuos_test.max()]
})

print("\n3. Estadísticas de Residuos:")
print(residuos_stats.round(4))

# Visualización de residuos vs predicciones
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(y_pred_test, residuos_test, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicciones')
plt.ylabel('Residuos')
plt.title('Residuos vs Predicciones')

plt.subplot(1, 2, 2)
sns.histplot(residuos_test, kde=True)
plt.title('Distribución de Residuos')
plt.xlabel('Residuos')
plt.tight_layout()
plt.show()
```

```python
print("\n4. CONCLUSIONES DEL MODELO:")
print("-"*50)
print("\na) Calidad del Modelo:")
if r2_score(y_test, y_pred_test) > 0.7:
    print("✓ El modelo tiene un buen rendimiento general")
else:
    print("⚠ El modelo podría necesitar mejoras")

print("\nb) Características más influyentes:")
for idx, row in feature_importance.head(3).iterrows():
    print(f"- {row['Feature']}: Impacto de {row['Importance']:.4f}")

print("\nc) Diagnóstico de Residuos:")
if abs(residuos_test.mean()) < 0.1 * residuos_test.std():
    print("✓ Los residuos están bien centrados alrededor de cero")
else:
    print("⚠ Posible sesgo en las predicciones")

print("\nd) Generalización del Modelo:")
r2_diff = abs(r2_score(y_train, y_pred_train) - r2_score(y_test, y_pred_test))
if r2_diff < 0.1:
    print("✓ Buena generalización entre training y test")
else:
    print("⚠ Posible sobreajuste")

print("\n5. RECOMENDACIONES:")
print("-"*50)
if r2_score(y_test, y_pred_test) < 0.7:
    print("- Considerar incluir más características relevantes")
    print("- Probar técnicas de regularización")
if r2_diff > 0.1:
    print("- Implementar técnicas de regularización")
    print("- Revisar la complejidad del modelo")
if abs(residuos_test.mean()) > 0.1 * residuos_test.std():
    print("- Revisar la distribución de los datos")
    print("- Considerar transformaciones no lineales")
```
